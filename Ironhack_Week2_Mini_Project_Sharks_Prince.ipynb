```python
# Split location into location and county
import pandas as pd
url = "https://www.sharkattackfile.net/spreadsheets/GSAF5.xls"

sharks_df = pd.read_excel(url)
sharks_df


# Split the 'Location' column into 'Location' and 'County'
sharks_df[['Location', 'County']] = sharks_df['Location'].str.split(',', n=1, expand=True)

# Clean up the new columns by stripping extra whitespace
sharks_df['Location'] = sharks_df['Location'].str.strip()
sharks_df['County'] = sharks_df['County'].str.strip()

sharks_df['Location'].unique()
```

```python

# Strip whitespace and standardize capitalization
sharks_df['Country'] = sharks_df['Country'].str.strip().str.title()
sharks_df['State'] = sharks_df['State'].str.strip().str.title()

# Replace common errors or inconsistent entries
country_replacements = {
    'Usa': 'USA', 'U.S.A.': 'USA', 'United States Of America': 'USA',
    'Aus': 'Australia', 'Uk': 'United Kingdom'
}

state_replacements = {
    'New South Wales ': 'New South Wales', 'Fla': 'Florida',
    'Ca': 'California', 'Wa': 'Washington'
}

sharks_df['Country'] = sharks_df['Country'].replace(country_replacements)
sharks_df['State'] = sharks_df['State'].replace(state_replacements)

# Fill missing values
sharks_df['Country'] = sharks_df['Country'].fillna('Unknown')
sharks_df['State'] = sharks_df['State'].fillna('Unknown')

# Check for unique values and outliers
print("Unique Countries:", sharks_df['Country'].unique())
print("Unique States:", sharks_df['State'].unique())

# Drop rows with invalid values if necessary
# Example: Drop rows with 'Unknown' in both columns
sharks_df = sharks_df[~((sharks_df['Country'] == 'Unknown') & (sharks_df['State'] == 'Unknown'))]

sharks_df[['Country','State']].head(40)
```
