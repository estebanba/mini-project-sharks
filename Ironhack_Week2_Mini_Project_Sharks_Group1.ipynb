```python
# def lower_case(x):
#     return x.lower()

# sharks_df["species"] = sharks_df["species"].fillna('').apply(lower_case)

# na_value = "N/A"

species_mapping_regex = {
    r'.*white.*': 'White Shark',
    r'.*tiger.*': 'Tiger Shark',
    r'.*bull.*': 'Bull Shark',
    r'.*wobbegong.*': 'Wobbegon Shark',
    r'.*nurse.*': 'Nurse Shark',
    r'.*blacktip.*': 'Blacktip Shark',
    r'.*blue.*': 'Blue Shark',
    r'.*raggedtooth.*': 'Raggedtooth Shark',
    r'.*mako.*': 'Mako Shark',
    r'.*lemon.*': 'Lemon Shark',
    r'.*zambesi.*': 'Zambesi Shark',
    r'.*sandshark.*': 'Sandshark',
    r'.*bronze whaler.*': 'Bronze Whaler Shark',
    r'.*caribbean reef.*': 'Caribbean Reef Shark',
    r'.*grey reef.*': 'Grey Reef Shark',
    r'.*dusky.*': 'Dusky Shark',
    r'.*horn.*': 'Horn Shark',
    r'.*sevengill.*': 'Sevengill Shark',
    r'.*carpet.*': 'Carpet Shark',
    r'.*galapagos.*': 'Galapagos Shark',
    r'.*porbeagle.*': 'Porbeagle Shark',
    r'.*hammerhead.*': 'Hammerhead Shark',
    r'.*copper.*': 'Copper Shark',
    r'.*leopard.*': 'Leopard Shark',
    r'.*shovelnose.*': 'Shovelnose Shark',
    r'.*spinner.*': 'Spinner Shark',
    r'.*whaler.*': 'Whaler Shark',
    r'.*bonita.*': 'Bonita Shark',
    r'.*angel.*': 'Angel Shark',
    r'.*thresher.*': 'Thresher Shark',

    r'.*questionable.*': "N/A",
    r'.*unconfirmed.*': "N/A",
    r'.*not confirmed.*': "N/A",
    r'.*invalid.*': "N/A",
     r'.*unknown.*': "N/A",

    r".*\s[a-z0-9\'\"\)\.\?\[\]]+$": "N/A",
    r'na[\'\"\)\.\?\[\]]+': "N/A",

    # Empty string and whitespace patterns
    r'^$': "N/A",               # Empty string
    r'^\s+$': "N/A",            # String containing only whitespace
    r'^["\']\s*["\']$': "N/A",  # Quoted empty string or whitespace
}

# sharks_df['species_clean'] = sharks_df['species'].str.lower().replace(species_mapping_regex, regex=True)

def clean_with_regex_dictionary(df, column, regex_map):
    df[column] = df[column].fillna("").apply(lambda x: x.lower())
    df[column] = df[column].str.lower().replace(regex_map, regex=True)
    return df

sharks_df = clean_with_regex_dictionary(sharks_df, "species", species_mapping_regex)

species_counts = sharks_df['species'].value_counts()

species_counts

# common_species = species_counts[species_counts >= 10].index.tolist()

# filtered_sharks_df = sharks_df[sharks_df['species_clean'].isin(common_species)]


```

```python
# def clean_fatal_rest(x):
#     if x not in ["N", "Y"]:
#         return "N/A"
#     else:
#         return x

# sharks_df['fatal_y/n_cleaned'] = sharks_df['fatal_y/n'].str.strip().str.upper().str.replace(r'[^a-zA-Z]', '', regex=True)
# sharks_df['fatal_y/n_cleaned'] = sharks_df['fatal_y/n_cleaned'].apply(clean_fatal_rest)

def clean_y_n(df, column):
    df[column] = df[column].str.strip().str.upper().str.replace(r'[^a-zA-Z]', '', regex=True)
    df[column] = df[column].apply(lambda x: "N/A" if x not in ["N", "Y"] else x)
    return df

sharks_df = clean_y_n(sharks_df, "fatal_y/n")


# sharks_df['fatal_y/n_cleaned'].value_counts()

sharks_df['fatal_y/n'].value_counts()
```

```python
# countries_counts = sharks_df['country'].value_counts()
# countries_counts

# filtered_df = sharks_df[(sharks_df['country'].isin(countries_top.keys())) & (sharks_df["species_clean"] != "N/A")]
# filtered_df.reset_index(inplace=True)
# filtered_df

def filter_column_by_min_count(df, column, min_count):
    value_counts = df[column].value_counts()
    filter_dict = value_counts[value_counts >= min_count]
    df = df[df[column].isin(filter_dict.keys())]
    df.reset_index(inplace=True)
    return df

def filter_column_by_value(df, column, value):
    df = df[df[column] != value]
    df.reset_index(inplace=True)
    return df

sharks_df = filter_column_by_min_count(sharks_df, "country", 50)
sharks_df = filter_column_by_value(sharks_df, "species", "N/A")

sharks_df
```
